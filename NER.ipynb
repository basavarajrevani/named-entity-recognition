```python
# Perform standard imports
import spacy
nlp = spacy.load('en_core_web_sm') #en_core_web_sm is a small English pipeline trained on written web text (blogs, news, comments), that includes vocabulary, syntax and entities.
```

```python
#Write a function to display basic entity info:
def show_ents(doc):
    if doc.ents:
        for ent in doc.ents:
            print(ent.text+' - ' +str(ent.start_char) +' - '+ str(ent.end_char) +
                  ' - '+ent.label_+ ' - '+str(spacy.explain(ent.label_)))
    else:
        print('No named entities found.')
```

```python

!pip install textblob
!python -m spacy download en_core_web_sm
import spacy
from spacy import displacy
from textblob import TextBlob

# Load the Spacy model (only once)
nlp = spacy.load('en_core_web_sm')

def show_ents(doc):
    if doc.ents:
        for ent in doc.ents:
            print(f'{ent.text} - {ent.start_char} - {ent.end_char} - {ent.label_} - {spacy.explain(ent.label_)}')
    else:
        print('No named entities found.')

def analyze_text(text):
    doc = nlp(text)

    # Named entities
    print("\nNamed Entities:")
    show_ents(doc)

    # Part-of-Speech Tagging and Dependency parsing
    print("\nPart-of-Speech Tagging and Dependency Parsing:")
    for token in doc:
        print(f"{token.text} - {token.pos_} - {token.dep_} - {token.head.text}")

    # Dependency parsing visualization
    print("\nDependency Parsing Visualization:")
    displacy.render(doc, style="dep", jupyter=True, options={'distance': 90})

    # Syntactic parsing visualization
    print("\nSyntactic Parsing Visualization:")
    displacy.render(doc, style="dep", jupyter=True, options={'distance': 90, 'compact': True})

    # Sentiment analysis using TextBlob
    sentiment = TextBlob(text).sentiment
    print(f"\nSentiment Analysis:")
    print(f"Polarity: {sentiment.polarity:.2f}")
    print(f"Subjectivity: {sentiment.subjectivity:.2f}")
    if sentiment.polarity > 0:
        print("Overall sentiment: Positive")
    elif sentiment.polarity < 0:
        print("Overall sentiment: Negative")
    else:
        print("Overall sentiment: Neutral")

# Example of user input
user_input = input("Enter a sentence or paragraph to analyze: ")

# Analyze the user input
analyze_text(user_input)
```
